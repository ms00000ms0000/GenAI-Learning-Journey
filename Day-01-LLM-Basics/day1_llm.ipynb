#### Import Required Libraries

```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
```

####  Load Environment Variables
 

```python
load_dotenv()
```

#### Configure Gemini API and Initialize Gemini Model

```python
genai.configure(api_key=os.getenv("Gemini API Keys"))
model=genai.GenerativeModel("gemini-2.5-flash")
```

#### Send Prompt to LLM and Display Generated Response

```python
response=model.generate_content("Explain machine learning to a 10-year-old child in three sentences.")
print(response.text)
```
