```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
load_dotenv()
genai.configure(api_key=os.getenv("Gemini API Keys"))
model=genai.GenerativeModel("gemini-2.5-flash")
response=model.generate_content("Explain machine learning to a 10-year-old child in three sentences.")
print(response.text)
```
