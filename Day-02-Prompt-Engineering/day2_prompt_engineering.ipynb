#### Import Required Libraries

```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
load_dotenv()
```

#### Gemini Model Initialization and Response Generation

```python
genai.configure(api_key=os.getenv("Gemini API Keys"))
model=genai.GenerativeModel("gemini-2.5-flash")

response=model.generate_content(" Explain Machine Learning  ")
print(response.text)

```

```python
response=model.generate_content(" Act as a professor, Explain Machine Learning with examples  ")
print(response.text)
```

```python
response=model.generate_content("You are a professor, Explain machine learning -Output Format- 1) Defintion, 2) Working 3) Example 4) Application. Explain in 200 words  ")
print(response.text)
```
